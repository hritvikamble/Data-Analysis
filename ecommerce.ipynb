# Q.2.1 DATA VISUALIZATION OF CATEGORIAL COLUMNS






```python
cat_columns=[]
for i in x.columns:
    
    if x[i].dtype=='object':
        print(i)
        cat_columns.append(i)
        print(x[i].nunique())
        if x[i].nunique()<60:
            print(x[i].value_counts())
        
        else:
            print('THIS COLUMN HAS MORE THAN 70 UNIQUE VALUES SO, GETTING VALUE COUNTS FOR THIS COLUMN IS NOT PREFFERABLE')
        print('__'*10)
```

# Q.3 Sales Trend Analysis

### Q.3.1 Group the data by month and calculate the total sales for each period

```python
i = '2023-09-30 00:42:28'
i

```

```python
i.split('-')[1]
```

```python
x['month order placed']=x['Order Date'].apply(lambda i:i.split('-')[1])
x
```

```python
monthlysales=x.groupby('month order placed')['Total Sales'].sum().reset_index()
monthlysales
```

### Q.3.2 Visualize sales trends overtime using bar plots

```python
plt.bar(monthlysales['month order placed'],monthlysales['Total Sales'],width=0.6,color='red')
plt.plot(monthlysales['month order placed'],monthlysales['Total Sales'],marker='*',lw=0.4,c='black')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('month vs sales')
plt.show()
```
